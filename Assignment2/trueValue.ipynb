```python
from Crypto.Cipher import DES
import binascii

# Function to convert hex string to bytes
def hex_to_bytes(hex_str):
    return binascii.unhexlify(hex_str)

# Function to perform DES encryption
def des_encrypt(plaintext_hex, key_hex):
    key = hex_to_bytes(key_hex)      # Convert key to bytes
    plaintext = hex_to_bytes(plaintext_hex)  # Convert plaintext to bytes

    cipher = DES.new(key, DES.MODE_ECB)  # Create DES cipher in ECB mode
    ciphertext = cipher.encrypt(plaintext)
    
    return binascii.hexlify(ciphertext).decode().upper()  # Convert to hex string

# Function to perform DES decryption
def des_decrypt(ciphertext_hex, key_hex):
    key = hex_to_bytes(key_hex)  # Convert key to bytes
    ciphertext = hex_to_bytes(ciphertext_hex)  # Convert ciphertext to bytes

    cipher = DES.new(key, DES.MODE_ECB)  # Create DES cipher in ECB mode
    decrypted_text = cipher.decrypt(ciphertext)
    
    return binascii.hexlify(decrypted_text).decode().upper()  # Convert to hex string

# Test cases
test_cases = [
    {'plaintext': '0123456789ABCDEF', 'key': '133457799BBCDFF1'},
    {'plaintext': 'FEDCBA9876543210', 'key': 'AABBCCDDEEFF0011'},
    {'plaintext': '0000000000000000', 'key': 'FFFFFFFFFFFFFFFF'}
]

# Encrypt and decrypt test cases
for test in test_cases:
    encrypted = des_encrypt(test['plaintext'], test['key'])
    decrypted = des_decrypt(encrypted, test['key'])

    print(f"Plaintext  : {test['plaintext']}")
    print(f"Key        : {test['key']}")
    print(f"Encrypted  : {encrypted}")
    print(f"Decrypted  : {decrypted}")
    print("-" * 40)
```
